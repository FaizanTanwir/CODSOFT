```python
X=df.drop(['Class'], axis=1)
Y=df[['Class']]
print(X.shape)
print(Y.shape)
```

```python
X.head(2)
```

```python
Y.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(X,Y,test_size=0.3,random_state=5,shuffle=True)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=25)
```

```python
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
pd.DataFrame(pred).head(5)
```

```python
pred_prb = model.predict_proba(xtest)[:,1]
```

```python
pd.DataFrame(pred_prb).head(5)
```

```python
xt = xtest.copy()
xt['pred_probability'] = pred_prb
xt['pred'] = pred
xt['actual'] = ytest
xt.head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest, pred)
```

```python
tn, fp, fn, tp = confusion_matrix(ytest, pred).ravel()
conf_matrix = pd.DataFrame({"predicted_fraud": [tp, fp], "pred_not fraud": [
                           fn, tn]}, index=["fraud", "not fraud"])
conf_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
recall = recall_score(ytest, pred)
print("Recall by built-in function: {}".format(recall))
```

```python
from sklearn.metrics import f1_score
```

```python
f1 = f1_score(ytest, pred)
print("F1 Score by built-in function: {}".format(f1))
```

```python
print(classification_report(ytest, pred))
```

```python
from sklearn.metrics import auc, roc_curve, roc_auc_score
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.figure(figsize=(8, 6))
    plt.title('ROC Curve')
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.005, 1, 0, 1.005])
    plt.xticks(np.arange(0, 1, 0.05), rotation=90)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend(loc='best')
```

```python
fpr, tpr, threshold = roc_curve(ytest, pred_prb)
```

```python
auc = roc_auc_score(ytest, pred_prb)
auc
```

```python
sns.set_context('poster')
plot_roc_curve(fpr, tpr, label='AUC = %0.3f' % auc)
```
